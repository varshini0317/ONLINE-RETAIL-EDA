```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['InvoiceYear'] = df['InvoiceDate'].dt.year
df['InvoiceMonth'] = df['InvoiceDate'].dt.month
```

```python
plt.figure(figsize=(8,6))
monthly_sales = df.groupby(['InvoiceYear', 'InvoiceMonth'])['Quantity'].sum()
monthly_sales.plot(marker='o')
plt.xlabel('Date')
plt.ylabel('Total Quantity Sold')
plt.title('Monthly Sales Trend')
plt.xticks(rotation=45)
plt.show()
```

```python
df['Amount'] = df['Quantity']*df['UnitPrice']
df.head()
```

```python
#CLUSTERING
```

```python
import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import cut_tree
```

```python
selected_features = ['Quantity', 'UnitPrice']
```

```python
scaler = StandardScaler()
df[selected_features] = scaler.fit_transform(df[selected_features])
```

```python
kmeans = KMeans(n_clusters=4, max_iter=50, random_state=42)
df['Cluster'] = kmeans.fit_predict(df[selected_features])
```

```python
df['Cluster_Id'] = kmeans.labels_
df.head()
```

```python
cluster_ids = df['Cluster_Id']
print(cluster_ids)
```

```python
sns.boxplot(x='Cluster_Id', y='Amount', data=df)
```

```python
# Visualize clusters
plt.scatter(df['Quantity'], df['UnitPrice'], c=df['Cluster'], cmap='viridis')
plt.xlabel('Quantity')
plt.ylabel('UnitPrice')
plt.title('K-Means Clustering Result')
plt.show()
```

```python
#KEY INSIGHTS:
#Customers with Cluster Id 0 are the customers with high amount of transactions as compared to other customers.
#Customers with Cluster Id 0 are frequent buyers.
```
